## 2. Variables et types de base


```python

# Variables numériques
age = 45
imc = 27.3

# Variable texte
sexe = "F"

# Variable booléenne
diabetique = True

print(age, imc, sexe, diabetique)
```


En Python :
- `int` → entier
- `float` → nombre réel
- `str` → texte
- `bool` → vrai / faux



## 3. Listes et dictionnaires


```python

# Liste (collection ordonnée)
ages = [23, 45, 67, 34]

# Dictionnaire (clé : valeur)
patient = {
    "age": 45,
    "imc": 27.3,
    "sexe": "F",
    "diabetique": True
}

print(ages)
print(patient)
```


## 4. Conditions


```python

age = 70

if age >= 65:
    print("Patient âgé")
else:
    print("Patient non âgé")
```


## 5. Boucles


```python

ages = [23, 45, 67, 34]

for age in ages:
    print(age)
```


## 6. Fonctions


```python

def risque_age(age):
    if age >= 65:
        return "Risque élevé"
    else:
        return "Risque modéré"

print(risque_age(45))
print(risque_age(72))
```


## 7. Introduction à NumPy


```python

import numpy as np

# Tableau numérique
data = np.array([120, 130, 110, 140])

print("Moyenne :", np.mean(data))
print("Écart-type :", np.std(data))
```


## 8. Introduction à Pandas


```python

import pandas as pd

# Création d'un DataFrame
df = pd.DataFrame({
    "age": [23, 45, 67, 34],
    "imc": [22.1, 27.3, 30.5, 24.8],
    "diabetique": [0, 1, 1, 0]
})

df
```


Quelques opérations courantes :
- `df.head()` → premières lignes
- `df.describe()` → statistiques
- `df["age"]` → accéder à une colonne



## 9. Visualisation simple


```python

import matplotlib.pyplot as plt

plt.hist(df["age"])
plt.xlabel("Âge")
plt.ylabel("Nombre de patients")
plt.title("Distribution des âges")
plt.show()
```
